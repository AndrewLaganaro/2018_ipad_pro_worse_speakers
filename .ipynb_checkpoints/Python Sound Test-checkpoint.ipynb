## Experiment Setup

I decided to record both iPads playing a [clip of this song](https://youtu.be/JQbjS0_ZfJ0) at full volume
* The iPad's were place right next to each other, with the microphone in the middle, approx 1 foot from the device
* The clip was played off of youtube for both devices and recorded using ableton software
* The duration of the clip was 17.5 seconds, starting from 1 minute in
* The clip was recorded using [this microphone](https://www.amazon.com/gp/product/B00HX0SRXW/ref=oh_aui_search_detailpage?ie=UTF8&psc=1)
* After recording I trimmed the clips so that they start at the exact same place and are the same length

## Verify audio is correct

I want to see if the bass frequencies of the old iPad are stronger than the new one. To do this I need to look at hte  [FFT](https://en.wikipedia.org/wiki/Fast_Fourier_transform) of both audio clips. This code creates a graph of both clips' FFT. To do this I followed [this great tutorial](https://pythondsp.rob-elder.com/loading-wav-files-and-showing-frequency-response/).


`Import required libraries`

```python
#%run ~/Apps/io_View/embed.ipynb
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fftpack import fft,fftfreq
```

`load in data, confirm audio is same length and from same part of song`

```python
samplerate, data_old = wavfile.read("old.wav")
samplerate, data_new = wavfile.read("new.wav")
data_old, data_new = data_old[:,0], data_new[:,0] #<-- there are 2 audio channels so we look at only one
```

```python
samples_old, samples_new = data_old.shape[0], data_new.shape[0] 
print(samples_old, samples_new)
```

```python
plt.subplot(2, 1, 1)
plt.plot(data_old, '-r' ) 
plt.title("Old iPad's Audio")
plt.subplot(2, 1, 2)
plt.plot(data_new ) 
plt.title("New iPad's Audio")
plt.tight_layout()
```

## View audio frequencies of iPads

`get fft data for both ipads`

```python
freqs = fftfreq(samples_old,1/samplerate)

datafft_old = fft(data_old)
fftabs_old = abs(datafft_old)

datafft_new = fft(data_new)
fftabs_new = abs(datafft_new)
```

`plot results`

```python
# Plot Old iPad Frequencies
plt.subplot(1, 2, 1) 
plt.xlim( [10, samplerate/2] )
plt.xscale( 'log' )
plt.grid( True )
plt.xlabel( 'Frequency (Hz)' )
plt.plot(freqs[:int(freqs.size/2)],fftabs_old[:int(freqs.size/2)]/7, '-r' , alpha=0.5)
plt.title("Old iPad Spectrum")

# Plot New iPad Frequencies
plt.subplot(1, 2, 2)
plt.xlim( [10, samplerate/2] )
plt.xscale( 'log' )
plt.grid( True )
plt.xlabel( 'Frequency (Hz)' )
plt.plot(freqs[:int(freqs.size/2)],fftabs_new[:int(freqs.size/2)] , alpha=0.5)
plt.title("New iPad's Spectrum")
plt.tight_layout()
```

## Experimental Results

The graphs above shows the (relative) level of noise produced by the 2 different iPads for all the audio frequencies on the x-axis. Frequencies to the left are "bassier" sounds.


 From looking at the 2 graphs it's pretty clear the the **old ipad has stronger bass frequencies**. The old  ipad could also reach a louder volume (not shown).
